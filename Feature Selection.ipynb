# Feature Selection Advance House Price Prediction

The main aim of this project to predict the House sale price based on various features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

##for feature selection
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

#to visualize all the columnns in the dataframe
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset = pd.read_csv('X_train.csv')
```

```python
dataset.head()
```

```python
#capture the dependent variable
y_train = dataset[['SalePrice']]
```

```python
#drop dependent variable from the dataset
X_train = dataset.drop(['Id','SalePrice'],axis=1)
```

```python
##apply feature selection
#first, I specify the Lasso regression model
#select a suitable alpha (equivalent of penalty)
#the bigger the alpha, the less feature that will be selected

#Then I use the selectFromModel object from sklearn which,
#will select the feature which coefficients are non zero

feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel_model.fit(X_train,y_train)
```

```python
feature_sel_model.get_support()
```

```python
#Lets print the number of total and selected features

#this is how we can make the list of selected features
selected_feat = X_train.columns[(feature_sel_model.get_support())]

#lets print the some stats
print("total feautes: {}".format((X_train.shape[1])))
print("Selected features: {}".format(len(selected_feat)))
print("features with coefficients shrank to zero: {}".format(np.sum(feature_sel_model.estimator_.coef_ == 0)))
```

```python
selected_feat
```

```python
X_train = X_train[selected_feat]
```
